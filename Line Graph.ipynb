```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
spotify_filepath = "spotify.csv"

# Read the file into a variable spotify_data
spotify = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)
spotify.head()
```

```python
spotify.tail()
```

```python
plt.figure(figsize=(20,10))
plt.title("Spotify Data")
sns.lineplot(data=spotify)
```

```python
list(spotify.columns)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=spotify['Shape of You'], label="Shape of You")
sns.lineplot(data=spotify['Despacito'], label="Despacito")
```

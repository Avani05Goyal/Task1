```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
Museum_path = "museum_visitors.csv"

# Read the file into a variable fifa_data
Museum = pd.read_csv(Museum_path, index_col="Date", parse_dates=True)
```

```python
Museum.head()
```

```python
Museum.tail()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=Museum)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(12,6))
# Add title
plt.title("Monthly Visitors to Avila Adobe")
# Line chart showing the number of visitors to Avila Adobe over time
sns.lineplot(data=Museum['Avila Adobe'])
# Add label for horizontal axis
plt.xlabel("Date")
```

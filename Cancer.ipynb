```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
cancer_b_path = "cancer_b.csv"
cancer_b = pd.read_csv(cancer_b_path, index_col="Id", parse_dates=True)
cancer_b.head()

```

```python
cancer_m_path = "cancer_m.csv"
cancer_m = pd.read_csv(cancer_m_path, index_col="Id", parse_dates=True)
cancer_m.head()


```

```python
sns.distplot(a=cancer_b['Area (mean)'],label='Benign' ,kde=False)
sns.distplot(a=cancer_m['Area (mean)'], label='malign',kde=False)

plt.legend()
```

```python
sns.kdeplot(data=cancer_b['Radius (worst)'], label="Cancer_b", shade=True)
sns.kdeplot(data=cancer_m['Radius (worst)'], label="Cancer_m", shade=True)
```
